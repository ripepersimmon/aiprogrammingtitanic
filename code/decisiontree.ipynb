```python
import pandas as pd
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
train_df = pd.read_csv('/workspaces/aiprogrammingtitanic/data/training.csv', index_col = 'PassengerId')
test_df = pd.read_csv('/workspaces/aiprogrammingtitanic/data/test.csv', index_col = 'PassengerId')
valid_df = pd.read_csv('/workspaces/aiprogrammingtitanic/data/validation.csv', index_col = 'PassengerId')

survived_train = train_df['Survived']

model = tree.DecisionTreeClassifier()

print(list(train_df))
filter = ['Sex', 'Fare','Age','Pclass','SibSp']
train_df = train_df[filter]
test_df = test_df[filter]
valid_df = valid_df[filter]

X_train = train_df.values
test = test_df.values
y_train = survived_train.values

model.fit(X_train, y_train)
```

```python
Y_pred = model.predict(test)
test_df['Survived'] = Y_pred
model.score(X_train, y_train)
```
